```python
#Data from https://data.world/etocco/nba-player-stats
#Data set of nba player stats per game from the 1997-98 season through 2021-22 season.

import pandas as pd
df3 = pd.read_csv('https://query.data.world/s/uljxcjekly2h7gn6ymimp2fqoxfcsk')
df3
```

```python
df3 = df3.dropna()

X = df3[['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']]

Y = df3[['MVP']]

x_names = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
## Models to be used: 
#Logistic Regression
#Random Forest
#K-nearest Neighbors
#Support Vector Machines
```

```python
#Logistic Regression
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot

model = LogisticRegression()
model.fit(X,Y) 
importance = model.coef_[0]
for i,v in enumerate(importance):
	print('Feature %0d: %s, Score: %.5f' % (i, x_names[i],v))

pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()

## Seems that GS, 3P, and PTS most positively affect MVP winners
## Seems that PF most negatively affects MVP winners (Honorable mention to 3PA)
```
